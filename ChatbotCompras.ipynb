```python
import mysql.connector
import telebot


bot = telebot.TeleBot("2016396929:AAGnKHlYNKIb4w8OJ0mmmBtH0vzG6lhCsmc")


db = mysql.connector.connect(
  host="MYSQL5027.site4now.net",
  user="a7b711_busproy",
  password="Sistemas10",
  port= "3306",  
  database= "db_a7b711_busproy"
)

cursor = db.cursor()

user_data = {}
class User:
    def __init__(self, first_name):
        self.first_name = first_name
        self.last_name = ''
        self.dni= ''
        self.origen= ''
        self.fecha= ''

@bot.message_handler(commands=['start', 'help'])
def send_welcome(message):
    msg = bot.send_message(message.chat.id, "Hola, ¿cual es su nombre?" )
    bot.register_next_step_handler(msg, process_firstname_step)   
    
def process_firstname_step(message):
    try:
        user_id = message.from_user.id
        user_data[user_id] = User(message.text)        
          
        msg = bot.send_message(message.chat.id, "¿Cual es su apellido?")
        bot.register_next_step_handler(msg, process_dni_step)
    except Exception as e:
        bot.reply_to(message, 'Ops dni problema"')

def process_dni_step(message):
    try:
        user_id = message.from_user.id
        user = user_data[user_id]
        user.last_name = message.text    
          
        msg = bot.send_message(message.chat.id, "¿Cual es su dni?")
        bot.register_next_step_handler(msg, process_origen_step)
    except Exception as e:
        bot.reply_to(message, 'Ops"')

def process_origen_step(message):
    try:
        user_id = message.from_user.id
        user = user_data[user_id]
        user.dni = message.text    
          
        msg = bot.send_message(message.chat.id, "¿Cual es su origen?")
        bot.register_next_step_handler(msg, process_fecha_step)
    except Exception as e:
        bot.reply_to(message, 'Ops"')
        
def process_fecha_step(message):
    try:
        user_id = message.from_user.id
        user = user_data[user_id]
        user.origen = message.text    
          
        msg = bot.send_message(message.chat.id, "¿Cual es su Fecha?")
        bot.register_next_step_handler(msg, process_final_step)
    except Exception as e:
        bot.reply_to(message, 'Ops"')
            
def process_final_step(message):
    try:
        user_id = message.from_user.id
        user = user_data[user_id]
        user.fecha = message.text
                
        sql = "INSERT INTO orders (nombres, apellidos, dni,origen,fecha) " \
               "VALUES (%s, %s, %s, %s, %s)"
        val = (user.first_name, user.last_name, user.dni,user.origen,user.fecha)
        cursor.execute(sql, val)
        db.commit()                
        
        msg = bot.send_message(message.chat.id, "Gracias por su registro")
        bot.register_next_step_handler(msg, process_consultar_step)
    except Exception as e:
        bot.reply_to(message, 'Ops, Segunda definicion')
        
def process_consultar_step(message):
    try:            
        cursor = db.cursor()
        cursor.execute("SELECT * FROM orders ")        
        vista = str(cursor.fetchall()[0:])                
        
        bot.send_message(message.chat.id, vista)
    except Exception as e:
        bot.reply_to(message, 'Ops, Tersera definicion')

bot.enable_save_next_step_handlers(delay=6)  
bot.load_next_step_handlers()
    
if __name__ == '__main__':
    bot.polling(none_stop=True)
```
